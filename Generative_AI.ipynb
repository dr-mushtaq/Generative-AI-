**Simple Python Example: Transformer-Based Text Generation**

Below is a minimal example using a transformer model from Hugging Face’s transformers library.

```python
from transformers import pipeline

# Load a pre-trained text generation model
generator = pipeline("text-generation", model="gpt2")

# Generate text
output = generator("Generative AI is transforming the world by", max_length=50)

print(output[0]["generated_text"])
```

**What’s happening here?**

A pretrained transformer model is loaded

It predicts the next word repeatedly

Self-attention helps maintain coherence

This simple example shows how transformer-based generative AI works in practice.
